```python
# Create a function to clean the data
def clean_network_data(df):
    """Clean network traffic data for ML"""
    print(f"Original data: {df.shape[0]:,} rows")
    
    # 1. Remove duplicates
    df_clean = df.drop_duplicates()
    print(f"After removing duplicates: {df_clean.shape[0]:,} rows ({df.shape[0] - df_clean.shape[0]:,} removed)")
    
    # 2. Replace infinite values with NaN
    df_clean = df_clean.replace([np.inf, -np.inf], np.nan)
    
    # 3. Fill missing values with median (better than mean for outliers)
    numeric_columns = df_clean.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        if df_clean[col].isnull().sum() > 0:
            median_val = df_clean[col].median()
            df_clean[col].fillna(median_val, inplace=True)
            print(f"Filled {col} missing values with median: {median_val}")
    
    # 4. Remove any remaining rows with NaN (if any)
    initial_count = df_clean.shape[0]
    df_clean = df_clean.dropna()
    if initial_count - df_clean.shape[0] > 0:
        print(f"Removed {initial_count - df_clean.shape[0]:,} rows with remaining NaN values")
    
    print(f"\n Final cleaned data: {df_clean.shape[0]:,} rows, {df_clean.shape[1]} columns")
    return df_clean

# Clean Tuesday's data
df_tuesday_clean = clean_network_data(df_tuesday)
```

```python
# Prepare data for machine learning
from sklearn.preprocessing import StandardScaler, LabelEncoder

print("PREPARING DATA FOR ML")
print("=" * 60)

# 1. Separate features (X) from labels (y)
X = df_tuesday_clean.drop([' Label'], axis=1)
y = df_tuesday_clean[' Label']

print(f"\n Features (X): {X.shape}")
print(f" Labels (y): {y.shape}")

# 2. Encode labels as numbers (BENIGN=0, attacks=1,2,3...)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"\n Label encoding:")
for i, label in enumerate(label_encoder.classes_):
    count = (y_encoded == i).sum()
    print(f"  {label}: {i} ({count:,} samples)")

# 3. Scale features to same range (0-1)
# This is CRITICAL - ML models need features on similar scales!
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"\n Features scaled using StandardScaler")
print(f"   Before scaling - Example values: {X.iloc[0, :3].values}")
print(f"   After scaling  - Example values: {X_scaled[0, :3]}")
```

```python
from sklearn.model_selection import train_test_split

# Split data: 70% training, 30% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, 
    test_size=0.3, 
    random_state=42,
    stratify=y_encoded  # Keep same proportion of attacks in train/test
)

print("DATA SPLIT")
print("=" * 60)
print(f"Training set: {X_train.shape[0]:,} samples")
print(f"Testing set:  {X_test.shape[0]:,} samples")

print(f"\n Training set distribution:")
unique, counts = np.unique(y_train, return_counts=True)
for label_code, count in zip(unique, counts):
    label_name = label_encoder.classes_[label_code]
    percentage = (count / len(y_train)) * 100
    print(f"  {label_name}: {count:,} ({percentage:.2f}%)")

print(f"\n Testing set distribution:")
unique, counts = np.unique(y_test, return_counts=True)
for label_code, count in zip(unique, counts):
    label_name = label_encoder.classes_[label_code]
    percentage = (count / len(y_test)) * 100
    print(f"  {label_name}: {count:,} ({percentage:.2f}%)")
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
import time

print("=" * 60)
print("TRAINING ISOLATION FOREST MODEL")
print("=" * 60)

# Create the model
# contamination = expected proportion of anomalies (we have ~2.2% attacks)
iso_forest = IsolationForest(
    contamination=0.022,  # 2.2% anomalies
    random_state=42,
    n_jobs=-1,  # Use all CPU cores
    verbose=1
)

# Train the model
print("\n Training model...")
start_time = time.time()
iso_forest.fit(X_train)
training_time = time.time() - start_time

print(f" Training completed in {training_time:.2f} seconds!")

# Make predictions
# Isolation Forest returns: 1 for normal, -1 for anomaly
print("\n Making predictions on test set...")
y_pred_iso = iso_forest.predict(X_test)

# Convert predictions: -1 (anomaly) -> 1, 1 (normal) -> 0
# So it matches our labels: 0=BENIGN, 1=attack, 2=attack
y_pred_binary = np.where(y_pred_iso == -1, 1, 0)
y_test_binary = np.where(y_test == 0, 0, 1)  # Convert to binary: 0=normal, 1=any attack

print("\n Predictions complete!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("=" * 60)
print("MODEL PERFORMANCE")
print("=" * 60)

# Calculate metrics
accuracy = accuracy_score(y_test_binary, y_pred_binary)
precision = precision_score(y_test_binary, y_pred_binary)
recall = recall_score(y_test_binary, y_pred_binary)
f1 = f1_score(y_test_binary, y_pred_binary)

print(f"\n Overall Metrics:")
print(f"  Accuracy:  {accuracy*100:.2f}% (how often is the model correct?)")
print(f"  Precision: {precision*100:.2f}% (when it says 'attack', how often is it right?)")
print(f"  Recall:    {recall*100:.2f}% (how many actual attacks did it find?)")
print(f"  F1-Score:  {f1*100:.2f}% (balance of precision and recall)")

# Confusion matrix
cm = confusion_matrix(y_test_binary, y_pred_binary)
print(f"\n Confusion Matrix:")
print(f"                  Predicted")
print(f"                Normal  Attack")
print(f"Actual Normal   {cm[0,0]:6,}  {cm[0,1]:6,}")
print(f"Actual Attack   {cm[1,0]:6,}  {cm[1,1]:6,}")

# Calculate detection rate for attacks
attacks_detected = cm[1,1]
total_attacks = cm[1,0] + cm[1,1]
detection_rate = (attacks_detected / total_attacks) * 100

print(f"\n Attack Detection Rate: {detection_rate:.2f}%")
print(f"   ({attacks_detected:,} out of {total_attacks:,} attacks detected)")
```

```python
# Get anomaly scores instead of binary predictions
# Lower score = more anomalous
anomaly_scores = iso_forest.score_samples(X_test)

print("ANOMALY SCORE ANALYSIS")
print("=" * 60)

# Check score distribution
print(f"\nAnomaly Score Statistics:")
print(f"  Min:    {anomaly_scores.min():.4f}")
print(f"  Max:    {anomaly_scores.max():.4f}")
print(f"  Mean:   {anomaly_scores.mean():.4f}")
print(f"  Median: {np.median(anomaly_scores):.4f}")

# Compare scores for normal vs attack traffic
normal_scores = anomaly_scores[y_test_binary == 0]
attack_scores = anomaly_scores[y_test_binary == 1]

print(f"\n Normal Traffic Scores:")
print(f"  Mean: {normal_scores.mean():.4f}")
print(f"  Std:  {normal_scores.std():.4f}")

print(f"\n Attack Traffic Scores:")
print(f"  Mean: {attack_scores.mean():.4f}")
print(f"  Std:  {attack_scores.std():.4f}")

print(f"\n Attacks have {'LOWER' if attack_scores.mean() < normal_scores.mean() else 'HIGHER'} scores on average")
print(f"   (Lower scores = more anomalous)")
```

```python
# Try different thresholds and see which works best
print("TESTING DIFFERENT THRESHOLDS")
print("=" * 60)

thresholds = [-0.45, -0.42, -0.40, -0.38, -0.36, -0.34]

best_f1 = 0
best_threshold = None
results = []

for threshold in thresholds:
    # Predict: if score < threshold, it's an attack
    y_pred_threshold = np.where(anomaly_scores < threshold, 1, 0)
    
    # Calculate metrics
    precision = precision_score(y_test_binary, y_pred_threshold, zero_division=0)
    recall = recall_score(y_test_binary, y_pred_threshold, zero_division=0)
    f1 = f1_score(y_test_binary, y_pred_threshold, zero_division=0)
    
    # Count detections
    cm = confusion_matrix(y_test_binary, y_pred_threshold)
    attacks_detected = cm[1,1] if cm.shape == (2,2) and len(cm[1]) > 1 else 0
    false_alarms = cm[0,1] if cm.shape == (2,2) and len(cm[0]) > 1 else 0
    
    results.append({
        'threshold': threshold,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'attacks_detected': attacks_detected,
        'false_alarms': false_alarms
    })
    
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

# Display results
print(f"\n{'Threshold':<12} {'Precision':<12} {'Recall':<10} {'F1-Score':<10} {'Attacks Detected':<18} {'False Alarms'}")
print("-" * 90)
for r in results:
    marker = " BEST" if r['threshold'] == best_threshold else ""
    print(f"{r['threshold']:<12.2f} {r['precision']*100:<11.1f}% {r['recall']*100:<9.1f}% {r['f1']*100:<9.1f}% {r['attacks_detected']:<18,} {r['false_alarms']:,} {marker}")

print(f"\n Best threshold: {best_threshold} (F1-Score: {best_f1*100:.2f}%)")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

print("=" * 60)
print("TRAINING RANDOM FOREST CLASSIFIER")
print("=" * 60)

# Create the model
rf_model = RandomForestClassifier(
    n_estimators=100,  # 100 decision trees
    max_depth=20,
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Train the model
print("\n Training model...")
start_time = time.time()
rf_model.fit(X_train, y_train)
training_time = time.time() - start_time

print(f" Training completed in {training_time:.2f} seconds!")

# Make predictions
print("\n Making predictions on test set...")
y_pred_rf = rf_model.predict(X_test)

print("\n Predictions complete!")
```

```python
print("=" * 60)
print("RANDOM FOREST PERFORMANCE")
print("=" * 60)

# Convert multi-class predictions to binary (0=normal, 1=attack)
y_pred_rf_binary = np.where(y_pred_rf == 0, 0, 1)

# Calculate metrics
accuracy = accuracy_score(y_test_binary, y_pred_rf_binary)
precision = precision_score(y_test_binary, y_pred_rf_binary)
recall = recall_score(y_test_binary, y_pred_rf_binary)
f1 = f1_score(y_test_binary, y_pred_rf_binary)

print(f"\n Overall Metrics:")
print(f"  Accuracy:  {accuracy*100:.2f}%")
print(f"  Precision: {precision*100:.2f}%")
print(f"  Recall:    {recall*100:.2f}%")
print(f"  F1-Score:  {f1*100:.2f}%")

# Detailed confusion matrix
cm = confusion_matrix(y_test_binary, y_pred_rf_binary)
print(f"\n Confusion Matrix:")
print(f"                  Predicted")
print(f"                Normal  Attack")
print(f"Actual Normal   {cm[0,0]:6,}  {cm[0,1]:6,} (False Alarms)")
print(f"Actual Attack   {cm[1,0]:6,}  {cm[1,1]:6,} (Correctly Detected)")

# Detection rate
attacks_detected = cm[1,1]
total_attacks = cm[1,0] + cm[1,1]
detection_rate = (attacks_detected / total_attacks) * 100

print(f"\n Attack Detection Rate: {detection_rate:.2f}%")
print(f"   Detected: {attacks_detected:,} out of {total_attacks:,} attacks")
print(f"   Missed: {cm[1,0]:,} attacks")
print(f"   False Alarms: {cm[0,1]:,}")

# Show detailed classification report
print(f"\n Detailed Classification Report:")
print(classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_))
```
